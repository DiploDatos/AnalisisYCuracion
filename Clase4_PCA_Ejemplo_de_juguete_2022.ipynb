# Clase 4: PCA: Ejemplo de juguete con el dataset Iris

En esta notebook, ejemplificaremos el uso del mtodo `PCA` para encontrar los componentes principales  y para reduccin de dimensionalidad.

Utilizaremos el conjunto de datos didctico [Iris](https://es.wikipedia.org/wiki/Conjunto_de_datos_flor_iris). Se trata de mediciones sobre la forma de distintos tipos de flores de iris y sus tipos respectivos. Ser muy adecuado para visualizar los componentes principales ya que las variables son combinaciones de distintas distribuciones gaussianas.

```python
import matplotlib.pyplot as plt
import numpy
import pandas

import seaborn
seaborn.set_context('talk')
```

```python
from sklearn import datasets

numpy.random.seed(5)

iris = datasets.load_iris()
X = iris.data
y = iris.target
X.shape, y.shape
```

Cargamos nuestro conjunto de datos Iris. La matriz de caractersticas tiene 150 filas y 4 columnas. Es decir:

$$X \in \mathbb{R}^{150 \times 4}$$

```python
X[:2]
```

```python
# To make visualizations more easily, we will construct a DataFrame with the
# matrix of values
columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
data_as_dict = {
      col_name: X[:,col_id] for col_id, col_name in enumerate(columns)
  }
data_as_dict['label'] = y.astype('str')
X_df = pandas.DataFrame(data=data_as_dict)
```

Primero analizaremos qu significa "proyectar" los datos a una nueva dimensin. La siguiente imagen muestra la proyeccin $\overrightarrow a_1 = (8, 0)$ del vector $ \overrightarrow a = (8, 5) $ sobre el eje horizontal.

![picture](https://vzahorui.net/assets/images/linear_algebra/vectors_and_cosine.png)


El mismo tipo de operacin se realiza sobre los vectores del conjunto de datos. Al visualizar dos columnas con un Scatterplot, cada vector $x_i \in \mathbb{R}^4$ (que corresponde a una fila), se representa slo como un vector de dos dimensiones. Se puede pensar que cada uno de estos puntos est siendo "proyectado" a una nica dimensin para graficar la distribucin de cada columna con los stripplots laterales. 

```python
g = seaborn.JointGrid(data=X_df, x='SepalLengthCm', y='SepalWidthCm',
                      height=10, hue='label')
g.plot_joint(seaborn.scatterplot, s=100)
g.plot_marginals(seaborn.stripplot)
```

Si se visualizan los datos a travs de Scatterplots de 2D, podemos ver cmo se distribuyen conjuntamente cada par de variables. El grfico `seaborn.pairplot` muestra tambin la distribucin de cada variable por separado utilizando histogramas.

Cada uno de estos grficos es una *proyeccin* de la matriz inicial a un nmero menor de dimensiones. En este caso, se utilizan las mismas dimensiones originales, es decir, las columnas pre-existentes, pero existen otras formas de proyeccin.




```python
seaborn.pairplot(data=X_df, hue='label')
```


# Reduccin de dimensionalidad

Para poder realizar un anlisis visual, nos limitaremos a trabajar en una nica dimensin. Es decir, se busca elegir una nica columna.

**Si el objetivo es predecir la columna `label`, qu columna nos ayuda a separar mejor nuestros datos?**



Eliminar todas las columnas permite reducir la dimensionalidad, pero como consecuencia se pierde mucha informacin. Existen mtodos que combinan la informacin de todas las columnas originales en una nica dimensin. Uno de ellos es Principal Component Analysis o **PCA**.

El PCA es un mtodo algebraico que nos permite calcular los componentes principales de una matriz. Los componentes principales son *direcciones* en el espacio de nuestros datos $R^{4}$ (ms especficamente son combinaciones lineales de los datos). Estos componentes o direcciones se calculan de forma tal que:
* Son ortogonales, es decir, no estn correlacionadas.
* Estn ordenados de acuerdo al nivel de varianza de los datos originales que representan.

Usando estos componentes, se puede construir una proyeccin lineal de nuestros datos a una nueva matriz Z donde cada columna ahora est en la direccin de un componente principal. Luego, se selecciona un subconjunto de las primeras $d$ columnas para formar la matriz $Z$. Por las propiedades de los componentes principales, sabemos que hemos perdido la menor cantidad de varianza de nuestros datos.





Utilizaremos la [clase PCA](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html) implementada en la librera scikit-learn. El mtodo `fit` computa los componentes principales de la matriz original, que luego podemos acceder a travs de los atributos de la instancia `pca`.


### Ejemplo 1: Proyeccin de $R^2$ en $R^2$ 

Con el propsito de visualizar la transformacin, utilizaremos slo dos columnas de la matrix X.

Al entrenar la instancia de la clase `PCA`, sklearn se encarga de centrar los datos restandoles la media. Sin embargo, es recomendable tambin *estandarizar* o al menos *escalar* la matriz original para asegurar de que todas las variables estn en las mismas unidades y ninguna tenga un peso demasiado grande.

```python
cols_to_project = ['PetalLengthCm', 'PetalWidthCm']
cols_to_project_ids = [columns.index(col) for col in cols_to_project]

X_standard = ((X - X.mean(axis=0)) / X.std(axis=0))[:,cols_to_project_ids]
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(X_standard)
print("Principal components")
print(pca.components_)
print("Explained variance: ")
print(pca.explained_variance_)
print("Explained variance ratio")
print(pca.explained_variance_ratio_)
```

En el siguiente grfico se representan los componentes principales. Se observa que son efectivamente las direcciones de mayor variacin.


```python
g = seaborn.JointGrid(x=X_standard[:,0], y=X_standard[:,1],
                      hue=X_df.label.values, height=10)
g.plot_joint(seaborn.scatterplot, s=100)
g.plot_marginals(seaborn.stripplot)
ax = g.ax_joint.figure.axes[0]
for length, vector in zip(pca.explained_variance_, pca.components_):
    arrowprops = dict(arrowstyle='->',
                      linewidth=2,
                      shrinkA=0, shrinkB=0)
    v = vector * numpy.sqrt(length)
    ax.annotate('', pca.mean_ + v, pca.mean_, arrowprops=arrowprops)
```

```python
X_projected = pca.transform(X_standard)

g = seaborn.JointGrid(x=X_projected[:,0], y=X_projected[:,1],
                      hue=X_df.label.values,
                      height=10)
g.plot_joint(seaborn.scatterplot, s=100)
g.plot_marginals(seaborn.stripplot)
```

### Ejemplo 2: Proyeccin de $\mathbb{R}^4$ en $\mathbb{R}^2$ 

En el ejemplo anterior, slo mostramos cmo los componentes principales permiten proyecta un conjunto de datos a *otras* dimensiones, pero utilizamos la misma cantidad.

Veamos ahora el resultado de calcular los componentes principales utilizando todas nuestras columnas, y luego cmo podemos proyectar esos datos de $\mathbb{R}^4$ a $\mathbb{R}^2$

```python
X_standard2 = ((X - X.mean(axis=0)) / X.std(axis=0))
```

```python
from sklearn.decomposition import PCA

pca2 = PCA(n_components=2)
pca2.fit(X_standard2)
print("Principal components")
print(pca2.components_)
print("Explained variance ratio")
print(pca2.explained_variance_ratio_)
```

```python
X_reduced = pca2.transform(X)

g = seaborn.JointGrid(x=X_reduced[:,0], y=X_reduced[:,1],
                      hue=X_df.label.values,
                      height=10)
g.plot_joint(seaborn.scatterplot, s=100)
g.plot_marginals(seaborn.stripplot)
```
